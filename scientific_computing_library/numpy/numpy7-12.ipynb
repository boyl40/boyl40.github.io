# 9 线性代数

Numpy可用于计算矩阵相乘、分解矩阵、求解线性方程等线性问题。

求解矩阵中的逆矩阵 

````python
import numpy.linalg as LA
````

````python
a = np.arange(1, 10, 1).reshape(3, 3)
````

````python
a
````

````python
# inva = LA.inv(a)
````

可以使用solve()函数解决下面这个线性方程组。
```
x1 + 2x2 = 2
2x1 - x2 = 4
```

````python
A = np.array([[1, 2], [2, -1]])
````

````python
A
````

````python
B = np.array([2, 4])
````

````python
LA.solve(A, B)
````

使用eig()函数求解一个矩阵的特征值和特征向量。 

````python
q = np.diag((4, 5, 6))
````

````python
q
````

````python
x, y = LA.eig(q)
````

````python
x
````

````python
y
````

# 11 Numpy的where()函数和统计函数 

## 11.1 where()函数

where()函数的语法是 where(condition, [x, y])。

````python
a = np.arange(5)
````

````python
a
````

````python
b = np.arange(5, 10)
````

````python
b
````

````python
c = np.array([True, False, True, False, True])
````

````python
c
````

````python
d = np.empty(5)
````

````python
d
````

````python
# 使用where条件(bool元素为True时，选择a, 否则选择b)
np.where(c, a, b)
````

````python
a = np.array([[2, -1], [-1, 2]])
````

````python
a
````

````python
np.where(a > 0, 1, 0)
````

# 11.2 统计函数 

````python
a = np.arange(9).reshape(3, 3)
````

````python
a
````

````python
a.sum()
````

````python
a.sum(axis = 0)  # 纵轴
````

````python
# 计算数组元素的平均值
a.mean()
````

````python
a.mean(0)  # 纵轴
````

````python
a.min(), a.min(0)
````

````python
a.max(), a.max(0)
````

````python
# 计算数组元素的标准差和方差
a.std(), a.var()
````

````python
# argmin、 argmax计算最小、最大元素的索引。
a.argmin(), a.argmax()
````

````python
# 计算元素的累加和
a.cumsum(axis = 0)
````
